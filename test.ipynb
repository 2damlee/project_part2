```python
from athleteData import Athlete, athletes
from gdpData import GdpData, gdp_data
from medalPayoutData import MedalPayoutData, medal_data

def test_read_athletes():
    file_path = 'data/athlete_events.csv'  # Replace with the path to your test CSV
    Athlete.read_athletes_csv(file_path)

    # Check the detail of the csv file(the first 5 rows)
    for idx, athlete in enumerate(athletes):
        if idx < 5:
            athlete.print_athlete_details()

    
    # Check if 'athletes' is a list
    assert isinstance(athletes, list), "Returned object should be a list"

    # Check if the list is not empty
    assert len(athletes) > 0, "Athlete list should not be empty"



def test_read_gdp():
    file_path = 'data/gdp_data.csv'
    GdpData.read_gdp_csv(file_path)
    
    # Check the detail of the csv file(the first 5 rows)
    for idx, data in enumerate(gdp_data):
        if idx < 5:
            data.print_gdp_details()
    
    # Check if 'gdp_data' is a list
    assert isinstance(gdp_data, list), "Returned object should be a list"
    
    # Check if the list is not empty
    assert len(gdp_data) > 0, "data list should not be empty"


def test_read_medalPayout():
    file_path = 'data/gold-medal-payout-by-country-2023.csv'
    MedalPayoutData.read_medalPayout_data_csv(file_path)
    
    # Check the detail of the csv file(the first 5 rows)
    for idx, data in enumerate(medal_data):
        if idx < 5:
            data.print_medal_details()
    
    # Check if 'medal_data is a lit
    assert isinstance(medal_data, list), "Returned object should be a list"
    
    # Check if the list is not empty
    assert len(medal_data) > 0, "data list should not be empty"
```
